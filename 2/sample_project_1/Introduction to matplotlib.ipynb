## Matplotlib example workflow

```python
# 0. import matplot lib and get it ready for ploting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot 
fig, ax = plt.subplots(figsize = (6, 10)) # (width, height)

# 3. plot data
ax.plot(x, y)

# 4. Customise plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. Save and show (you save the whole figure)
fig.savefig("images/sample_plot.png")
```
