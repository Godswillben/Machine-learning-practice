## Matplotlib example workflow

```python
# 0. import matplot lib and get it ready for ploting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot 
fig, ax = plt.subplots(figsize = (6, 10)) # (width, height)

# 3. plot data
ax.plot(x, y)

# 4. Customise plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. Save and show (you save the whole figure)
fig.savefig("images/sample_plot.png")
```

### Which one should be used? (pyplot vs matplotlib OO method)
* When plotting somthing quickly okay to use pyplot method
* When plotting something more advanced use the OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
over_50.plot(kind="scatter", x="age", y="chol", c='target')
```

```python
#00 method
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter', x='age', y='chol', c="target", ax=ax)

ax.set_xlim([45, 100])
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])

# customize the plot
ax.set(title="Heart Disease and cholesterol Levels", xlabel="Age", ylabel="Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")

# Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle="--")
```

```python
# Subplots of age, chol, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"], y=over_50['chol'], c=over_50['target'])

# customise ax0
ax0.set(title="Heart diseases and cholesterol levels", xlabel="age", ylabel="cholesterol")

# Add legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# Add meanline
ax0.axhline(y=over_50["chol"].mean(), linestyle="--")

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"])

# Customize ax1
ax1.set(title="Heart Disease and Max heart rate", xlabel='Age', ylabel="Max heart rate");

# add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--")

# Add title to the figure
fig.suptitle("Heart Disease analysis", fontsize=16, fontweight="bold");
```
